- calculate the frequency of all 26 characters

```python
import nltk
```

```python
#nltk.corpus.gutenberg.fileids()
```

```python
#reading the raw file
raw_file = nltk.corpus.gutenberg.raw('melville-moby_dick.txt') 
#raw_file
```

```python
#converting lines to lower case

lines = raw_file.lower()
#lines
```

```python
#using counter

#from collections import Counter
#counts = Counter(c for c in lines if c.isalpha()) #getting a count of all the words in the file
#print(counts)
```

```python
#creating a dictionary of alphabates with their frequency
alpha = {}

for a in lines:
    if a.isalpha():
        if a in alpha:
            alpha[a] +=1
        else:
             alpha[a] = 1
```

```python
#sorting the dictionary
import operator

sorted_list = sorted(alpha.items(),key=operator.itemgetter(1),reverse=True)
```

```python
#printing all the characters based on their frequency
for word in sorted_list:
    print(word[1], word[0])
```
