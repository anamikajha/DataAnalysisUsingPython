```python
import nltk
nltk.corpus.names.fileids()
```

```python
female_names = nltk.corpus.names.words('female.txt')
male_names = nltk.corpus.names.words('male.txt')
```

```python
#creating a list of all the names
total_names = []

for name in female_names:
    total_names.append(name)
    
for name in male_names:
    total_names.append(name)
```

```python
#checking if we have common names in both in list

print('total no of names: ', len(total_names))
print('total no of unique names: ',len (set(total_names)))
```

```python
#creating a dictionary with all the names along with its frequency
total = {}

for name in female_names:
        total[name] = 1
        
for n in male_names:
    if n in total:
        total[n]+= 1
    else:
        total[n] = 1
```
