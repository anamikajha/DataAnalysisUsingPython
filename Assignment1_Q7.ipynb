# Question 7
- Read the data from Name corpus
- Calculate the probability of a name being a male or female name based on last character of the name

```python
import nltk
#nltk.corpus.names.fileids()
```

```python
female_names = nltk.corpus.names.words('female.txt')
male_names = nltk.corpus.names.words('male.txt')
```

```python
#defining functions to caluculate the probabilities

def func_male_prob(a,b):
    try:
        male_prob = a/b * 100
        return male_prob
    except:
        return "No Value found"
    
def func_female_prob(a,b):
    try:
        female_prob = a/b * 100
        return female_prob
    except:
        return "No Value found"
```

```python
#calculating probabilities

#creating a set of characters
from string import ascii_lowercase


for a in ascii_lowercase:
    total_names = []
    total_female_names = []
    total_male_names = []
    for word in female_names:
        if word.endswith(a):
            total_names.append(word)
            total_female_names.append(word)

    for word in male_names:
        if word.endswith(a):
            total_names.append(word)
            total_male_names.append(word)
    
    male_prob = func_male_prob(len(total_male_names),len(total_names))
    female_prob = func_female_prob(len(total_female_names),len(total_names))
    
    if male_prob != 'No Value found':
        print('[',a,']  male:',male_prob,'%, female:',female_prob,'%')
    else:
        print('[',a,'] No Values found ending in',a)
```
