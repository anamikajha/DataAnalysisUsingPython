# Question 2 - Part 1

- Reading all json files from Data folder.
- Create different folders to store these files based on their geography
- Hierarchical folder structure created to move the files (Country - State - City)

```python
#use of glob function to read the file
import glob
import shutil
import os
import json
json_files = glob.glob('Data/*.json') #absolute path to read the jason files
for file in json_files:
    with open(file) as data_file:
        data = json.load(data_file)
        country=data['location']['country'] #reading the country name from each jason file
        #print (country)
        state=data['location']['state'] #reading the state name from each jason file
        city=data['location']['city']   #reading the city name from each jason file
        if not os.path.exists('Data Processed/'+country+'/'+state+'/'+city):
            os.makedirs('Data Processed/'+country+'/'+state+'/'+city)
            #creating a hierarchy of folders as country-state-city to move the proper json file
            shutil.copyfile('Data/'+file[5:],'Data Processed/'+country+'/'+state+'/'+city+'/'+file[5:])
```
