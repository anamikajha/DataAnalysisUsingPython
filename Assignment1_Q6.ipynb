# Question 6
- Get data from Inaugural address package
- Get unique words used by every president


```python
import nltk
files = nltk.corpus.inaugural.fileids()
```

```python
#Getting the names of president 
file_names = []
for f in files:
    file_names.append(f[5:-4])

file_names
```

```python
#combining the lists of same presidents
for f in file_names:
    for file in files:
        #print(f)
        if file[5:-4] == f:
            word_list = []
            words = set(nltk.corpus.inaugural.words(file))
            word_list.extend(words)
            print ('Set of unique words in' ,f, 'speech : ', len(word_list))
            
    #words= set(nltk.corpus.inaugural.words(f))
    #print ('Set of unique words in' ,f[5:-4], ': ', len(words))
```
