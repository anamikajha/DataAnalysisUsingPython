# Analyzing the reason behind pairing of questions
- Studying the evolution of question over time (if they are same)
- Analyzing why they were paired together even if they are not duplicate questions
- calculating the matching index of the non duplicate questions
- If the matching index is 100% analyze if one question is a subset of other


```python
#getting all the required imports
import numpy as np
import pandas as pd
import re
import seaborn as sns
import matplotlib.pyplot as plt
from subprocess import check_output

%matplotlib inline
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls
```

```python
import os
#getting current working directory
b = os.getcwd()

df = pd.read_csv(b+"/"+"Data/questions.csv", usecols=[3,4,5]).fillna("")
df.head() 
```

```python
#filtering out the rows that are not duplicate
unmatched_df = df[df['is_duplicate'] == 0]
```

```python
#resetting the index to create a continuous row ids
unmatched_df = unmatched_df.reset_index()
```

```python
# Offsetting the warning
pd.options.mode.chained_assignment = None

from fuzzywuzzy import fuzz

#calculate the matching index of the two questions using fuzzywuzzy
indexs = []
for i in range(0,len(unmatched_df)):
    indexs.append(fuzz.token_set_ratio(unmatched_df['question1'][i],unmatched_df['question2'][i]))                                         
```

```python
#converting the list to dataframe so that it can be appended to the existing df
indexs = pd.DataFrame(indexs)
```

```python
#concatenating the dataframe with matching index column
matching_df = pd.concat([unmatched_df, indexs], axis=1)
matching_df.columns = ['index','question1','question2','is_duplicate','matching index']
```

```python
matching_df.head(7)
```

```python
#extracting out the question pair where matching index value comes to 100%
similar_q = matching_df[matching_df['matching index']== 100]
```

```python
similar_q.head()
```

```python
#resetting the index to create a continuous row ids
similar_q = similar_q.reset_index()
```

```python
#deleting unwanted columns
del similar_q['level_0']
del similar_q['index']
```

```python
similar_q.head()
```

```python
#creating a blank dataset
is_subset = ['']
is_subset = pd.DataFrame(is_subset)
```

```python
#concatenating dataframe with is subset column
is_subset_df = pd.concat([similar_q, is_subset], axis=1)
is_subset_df.columns = ['question1','question2','is_duplicate','matching index','is subset']
```

```python
#checking if Question1 is a subset of Question2
is_subset_1 = []
is_subset_2 = []
for i in range(0,len(is_subset_df)):
    if (is_subset_df['question1'][i] in is_subset_df['question2'][i]):
        is_subset_df['is subset'] = 'Y'
    elif (is_subset_df['question2'][i] in is_subset_df['question1'][i]):
        is_subset_df['is subset'] = 'Y'
```

```python
#total number of questions in the subset dataframe
len(is_subset_df)
```

```python
#number of rows where the questions are subset 
len(is_subset_df[is_subset_df['is subset']== 'Y'])
```

- Since the length of both the dataframes are same, Hence we see that all the questions with a matching index of 100 were a subeset of the other question it is paired with. This explains why they were assigned a matching index of 100.
