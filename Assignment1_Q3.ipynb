```python
import nltk
```

```python
nltk.corpus.webtext.fileids()
```

```python
#getting the raw content from the file
raw_file = nltk.corpus.webtext.raw('pirates.txt')
#raw_file
```

```python
#getting count of lines by both the characters
jack_count = raw_file.count('\nJACK SPARROW:')
will_count = raw_file.count('\nWILL TURNER:')
```

```python
print(jack_count)
print(will_count)

if jack_count > will_count:
    print('Jack Sparrow is the leading man')
else:
    print('Will Turner is the leading man')
```
