```python
for index, row in df.iterrows():
    if(pd.isnull(df['VEHICLE 2 TYPE']) & pd.isnull(df['VEHICLE 3 TYPE']) & pd.isnull(df['VEHICLE 4 TYPE']) & pd.isnull(df['VEHICLE 5 TYPE'])):
        df['ONE_VEHICLE_INVOLVED'] = 1
    elif((pd.isnull(df['VEHICLE 3 TYPE'])== True) & (pd.isnull(df['VEHICLE 4 TYPE'])== True) & (pd.isnull(df['VEHICLE 5 TYPE'])== True)):
        df['TWO_VEHICLE_INVOLVED'] = 1
    elif((pd.isnull(df['VEHICLE 4 TYPE'])== True) & (pd.isnull(df['VEHICLE 5 TYPE'])== True)):
        df['THREE_VEHICLE_INVOLVED'] = 1
    else:
        df['MANY_VEHICLE_INVOLVED'] = 1
```
