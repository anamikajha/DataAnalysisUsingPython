- calculate the frequency for each words
- top 10 most frequent words (can include punctuations)

```python
import nltk
```

```python
from nltk.corpus import gutenberg
```

```python
#nltk.corpus.gutenberg.fileids()
```

```python
#getting the total no of words
words = nltk.corpus.gutenberg.words('austen-emma.txt') 
```

```python
#converting upper case to lower
words_lower = []
for x in words:
    w = x.lower()
    words_lower.append(w)

#words_lower
```

```python
#calculating the total no of words in the file
len(words_lower)
```

```python
#calculating number of unique words
len (set(words_lower))
```

```python
#creating a dictionary of words and their frequency
counter = {}
for word in words_lower:
    if word in counter:
        counter[word] +=1
    else:
         counter[word] = 1
```

```python
#Printing the frequency of all the words
counter
#type(counter)
```
